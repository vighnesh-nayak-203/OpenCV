```python
import numpy as np
import cv2 as cv
import pandas as pd
import matplotlib.pyplot as plt
from keras import Sequential
from keras.layers.core import Dense,Flatten
from keras.layers import Convolution2D,MaxPooling2D
from sklearn.model_selection import train_test_split as split
from keras.utils import to_categorical
from keras import backend as K
import keras.datasets.mnist as mnist
```

```python
data=mnist.load_data()
```

```python
X=data[0][0]
Y=data[0][1]
```

```python
Y=to_categorical(Y,num_classes=10)
```

```python
print(X.shape,Y.shape)
```

```python
x_train,x_test,y_train,y_test=split(X,Y,test_size=0.15,random_state=4)
```

```python
model=Sequential()
model.add(Flatten())
model.add(Dense(256,'relu'))
model.add(Dense(256,'relu'))
model.add(Dense(10,'softmax'))
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
model.build((None,28,28))
```

```python
model.summary()
```

```python
his=model.fit(x_train,y_train,
              validation_split=0.15,
              epochs=10,
             )
```

```python
model1=Sequential()
model1.add(Convolution2D(32,5,1,input_shape=(28,28,1),activation='relu'))
model1.add(Convolution2D(16,5,1,input_shape=(20,20,16),activation='relu'))
model1.add(MaxPooling2D((2,2)))
model1.add(Flatten())
model1.add(Dense(256,'relu'))
model1.add(Dense(256,'relu'))
model1.add(Dense(10,'softmax'))
model1.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
model1.build((None,28,28))
```

```python
model1.summary()
```

```python
his1=model1.fit(x_train,y_train,
              validation_split=0.15,
              epochs=2,
             )
```

```python
score = model.evaluate(x_test, y_test, verbose = 0) 

print('Test loss:', score[0]) 
print('Test accuracy:', score[1])
```

```python
score1 = model1.evaluate(x_test, y_test, verbose = 0) 

print('Test loss:', score1[0]) 
print('Test accuracy:', score1[1])
```
